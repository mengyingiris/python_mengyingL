```python
#importing the library as shortname 
import matplotlib.pyplot as plt 
```

```python
photos_inventory = {
    'New York': 4267, 
    'Illinois': 7000, 
    'California': 1400, 
    'Hawaii': 811, 
    'Washington': 890, 
    'Florida': 627,
    }
```

```python
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html
plt.bar(range(len(photos_inventory)), list(photos_inventory.values()), align='center', color=['#a5b2c7', '#87cbed', '#f5cb62', '#f29b22', '#5ea851', '#72e8cc'])
plt.xticks(range(len(photos_inventory)), list(photos_inventory.keys()), rotation = 45)
plt.show()
```

```python
#https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter
plt.scatter(range(len(photos_inventory)), list(photos_inventory.values()), color=['#a5b2c7', '#87cbed', '#f5cb62', '#f29b22', '#5ea851', '#72e8cc'])
plt.xticks(range(len(photos_inventory)), list(photos_inventory.keys()), rotation = 45)
plt.show()
```

```python
import matplotlib.pyplot as plt

slices = [4267,7000,1400,811,890,627]
activities = ['NY','IL','CA','HI','WA','FL']
cols = ['#a5b2c7', '#87cbed', '#f5cb62', '#f29b22', '#5ea851', '#72e8cc']
plt.pie(slices,
labels=activities,
colors=cols,
startangle=40,
shadow= True,
#To explode a pie chart means to make one of the wedges of the pie chart to stand out. 
explode=(0,0.1,0,0,0,0),
autopct='%1.1f%%') # to the nearest .1% float 
plt.title('US Photos in Pie Plot')
```
